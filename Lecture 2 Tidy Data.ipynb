```python
my_aqi_file=open('aqi_messy.csv','r+') #it assigns whatever is in 'aqi_messy' to a my_aqi_file object
aqifile_list=my_aqi_file.readlines() #let's create a list reading the lines of the my_aqi_file object
my_aqi_file.close() # and let's close the file, since we have finished
```

```python
for a in range (0,4): 
    aqifile_list.pop(0)
```

```python
aqifile_list
```

```python
aqifile_list_tidy=[]#we prepare our tidy list, empty. We will fill it in soon!

for counter in range (0,len(aqifile_list)): #do something a number of times equal to the length of the list
    aqifile_list_tidy.append(aqifile_list[counter].split(' ')) #that is: add to the new tidy table a line with 
                                                        # 3 new items: the line of the old table split at the character
                                                        # 'space'
```

```python
aqifile_list_tidy #oh look, much better! but wait; we have an empty first column which we don't need
```

```python
for counter in range (0,len(aqifile_list_tidy)): #do something a number of times equal to the length of the new list
    aqifile_list_tidy[counter].pop(2) #remove the third item in each line (-->the third column in each mini-list)
    aqifile_list_tidy[counter].pop(0) #remove the first item in each line (-->the first column in each mini-list)
                                    
```

```python
aqifile_list_tidy #ok, now the empty columns are gone. We have other issues though... 
                    #look at the third column! is that a number?
```

```python
for counter in range (0,len(aqifile_list_tidy)): #do something a number of times equal to the length of the new list
    aqifile_list_tidy[counter][2]=aqifile_list_tidy[counter][2][1:]  #take everything (:) from the second character 
                                                                        #onward (1)
    aqifile_list_tidy[counter][2]=aqifile_list_tidy[counter][2].split("\n")[0] #split the cell at "\" and take 
                                                                                #the left part

aqifile_list_tidy #alright, much better. Now we have a number in the final column.
```

```python
for counter in range (1,len(aqifile_list_tidy)):  #why 1? we don't want the headers to become numbers (they wouldn't -->error)
       # print counter #let's see where the problem lies
        if aqifile_list_tidy[counter][2]=='':
            aqifile_list_tidy[counter][2]='0.00'
        aqifile_list_tidy[counter][2]=float(aqifile_list_tidy[counter][2]) #and now that we removed all nuisances convert that string into a float

```

```python
aqifile_list_tidy[0:10] #alright, so much better! Now it's tidy enough
```

```python
aqi_dataframe=pd.DataFrame(aqifile_list_tidy[1:], columns=aqifile_list_tidy[0]) #ok, let's bring it back to Pandas
                                                                                #taking everything except the first line
                                                                                #[1:]
                                                                                #
```

```python
aqi_dataframe.head() #there! a tidy dataset
aqi_dataframe.columns=['Date','Time','NO2']
```

```python
import matplotlib.pyplot as plt

aqi_to_plot=aqi_dataframe.groupby(['Date'],as_index=False).mean()
aqi_to_plot.head()

```

```python
#plt.plot_date (aqi_dataframe['Date'],aqi_dataframe['NO2'],'g-')
plt.plot_date (aqi_to_plot['Date'],aqi_to_plot['NO2'],'g-')

plt.show()
```

```python
plt.rc('xtick', labelsize=8) 
plt.rc('ytick', labelsize=8) 
plt.rcParams["axes.labelsize"] = 8

mygraph=plt.plot_date (aqi_to_plot['Date'],aqi_to_plot['NO2'],'g-',label="NO2 mg/M3")
plt.legend(handles=mygraph)

plt.show()
```
